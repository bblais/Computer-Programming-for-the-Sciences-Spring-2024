```python
%matplotlib inline
from pylab import *
from pyndamics3 import Simulation
import pandas as pd
```

```python
data=pd.read_csv('logistic_sample_data/logistic_sample_data_0.csv')
data
```

```python
t_data=data['t'].values
y_data=data['y'].values
t_data,y_data
```

```python
sim=Simulation()
sim.add("y' = r*y*(1-y/K)",initial_value=1)
sim.params(r=1,K=10)
sim.run(50)
```

```python
plot(sim.t,sim.y)
plot(t_data,y_data,'-o')
xlabel('time')
ylabel('population')
```

```python
sim=Simulation()
sim.add("y' = r*y*(1-y/K)",initial_value=1)
sim.params(r=1,K=100)
sim.run(50)

plot(sim.t,sim.y)
plot(t_data,y_data,'-o')
xlabel('time')
ylabel('population')
```

```python
sim=Simulation()
sim.add("p' = r*p*(1-p/K)",initial_value=1)
sim.params(r=1,K=250)
sim.run(50)

plot(sim.t,sim.y)
plot(t_data,y_data,'-o')
xlabel('time')
ylabel('population')
```
