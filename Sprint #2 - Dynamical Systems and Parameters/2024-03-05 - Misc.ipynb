```python
%matplotlib inline
from pylab import *
from pyndamics3 import Simulation
```

```python
sim=Simulation()
sim.figsize=(6,3)
sim.add("y'=a",1,plot=True)
sim.params(a=2)
sim.run(10)
```

Let's say the constant changes midway

```python
def a(t,t1):
    if t<t1:
        return 2
    else:
        return 4


sim=Simulation()
sim.figsize=(6,3)
sim.add("y'=a(t,t1)",1,plot=True)
sim.params(t1=5)
sim.functions(a)
sim.run(10)
```
