```python
from pyndamics3 import Simulation
```

## Logistic growth

$$
\frac{dx}{dt} = ax (1-x/K)
$$

```python
sim=Simulation()
sim.figsize=(8,4)   # you dont need this line -- just for the projector
sim.add(" x' = a*x*(1-x/K) ",initial_value=2,plot=True)
sim.params(a=1,K=50)
sim.run(10)

```

## Free fall model

$$
\frac{dx}{dt}= v
$$

$$
\frac{dv}{dt}=-g + k v^2/m
$$

```python
sim=Simulation()
sim.figsize=(8,4)   # you dont need this line -- just for the projector
sim.add(" x' = v ",initial_value=1000,plot=True)
sim.add(" v' = -g + k*v**2/m ",initial_value=0,plot=True)
sim.params(k=1,g=10,m=1)
sim.run(325)  # the number here is the amount of time it runs
```

## Parameter Exploration

### Method one

This method is a bit of a hack but is easy.

```python
sim=Simulation()
sim.figsize=(8,4)   # you dont need this line -- just for the projector
sim.add(" x1' = a*x1*(1-x1/K1) ",initial_value=2,plot=1)  # instead of True use a figure number
sim.add(" x2' = a*x2*(1-x2/K2) ",initial_value=6,plot=1)
sim.params(a=1,K1=50,K2=100)
sim.run(10)

```

### Method 2

This gives you more control and is the way to properly do it.

```python
sim=Simulation()
sim.figsize=(8,4)   # you dont need this line -- just for the projector
sim.add(" x' = a*x*(1-x/K) ",initial_value=2,plot=False)
sim.params(a=1,K=50)
sim.run(10)

x1=sim.x
t1=sim.t

sim=Simulation()
sim.figsize=(8,4)   # you dont need this line -- just for the projector
sim.add(" x' = a*x*(1-x/K) ",initial_value=2,plot=False)
sim.params(a=1,K=100)
sim.run(10)

x2=sim.x
t2=sim.t

plot(t1,x1)
plot(t2,x2)
```

```python
for K in [10,50,100]:
    sim=Simulation()
    sim.figsize=(8,4)   # you dont need this line -- just for the projector
    sim.add(" x' = a*x*(1-x/K) ",initial_value=2,plot=False)
    sim.params(a=1,K=K)
    sim.run(10)
    
    x=sim.x
    t=sim.t

    plot(t,x,label=f"K={K}")

xlabel('Time')
ylabel('Population')
legend()
```

```python
for k in [1,5,10]:
    sim=Simulation()
    sim.figsize=(8,4)   # you dont need this line -- just for the projector
    sim.add(" x' = v ",initial_value=1000,plot=False)
    sim.add(" v' = -g + k*v**2/m ",initial_value=0,plot=False)
    sim.params(k=k,g=10,m=1)
    sim.run(10)  # the number here is the amount of time it runs


    figure(1)
    plot(sim.t,sim.x)

    figure(2)
    plot(sim.t,sim.v)
```
