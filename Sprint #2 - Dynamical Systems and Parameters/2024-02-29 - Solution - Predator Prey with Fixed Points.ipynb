```python
%matplotlib inline
from pylab import *
from pyndamics3 import Simulation
```

```python
sim=Simulation()
sim.add(" x' = *x - *x*y",10,plot=True)
sim.add(" y' = *x*y-*y",10,plot=True)
sim.params(=1.1 ,=0.4, =.1, =.4 )
sim.run(100)
```

```python
sim=Simulation()
sim.add(" x' = *x - *x*y",4.1)
sim.add(" y' = *x*y-*y",3)
sim.params(=1.1 ,=0.4, =.1, =.4 )
sim.run(100)

figure(figsize=(8,4))
plot(sim.t,sim.x)
plot(sim.t,sim.y)

plot(sim.t,0*ones(len(sim.t)),'r--')

# prey
plot(sim.t,(.4/.1)*ones(len(sim.t)),'c--')

# predator
plot(sim.t,(1.1/.4)*ones(len(sim.t)),'r--')
```

```python
sim=Simulation()
sim.add(" x' = *x - *x*y",4.1)
sim.add(" y' = *x*y-*y",3)
sim.params(=1.1 ,=0.4, =.1, =.4 )
sim.run(100)

figure(figsize=(8,4))
plot(sim.t,sim.x)
plot(sim.t,sim.y)

plot(sim.t,0*ones(len(sim.t)),'r--')

# prey
plot(sim.t,(sim./sim.)*ones(len(sim.t)),'c--')

# predator
plot(sim.t,(1.1/.4)*ones(len(sim.t)),'r--')
```
