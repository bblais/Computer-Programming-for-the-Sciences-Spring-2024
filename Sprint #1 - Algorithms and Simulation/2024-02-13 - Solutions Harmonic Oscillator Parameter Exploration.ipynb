$$
\frac{dx}{dt}= v 
$$

$$
\frac{dv}{dt}= -k\cdot x/m
$$

## set initial conditions for variables

```python
x=3
v=0
t=0
dt=0.01
```

## set values of parameters

```python
k=1
m=2
```

## loop the model equations

- calculate small changes in the variables
- update variables

```python
S=Storage()

S+=t,x,v

while t<10:
    
    dx = (v) *dt  # small change in x
    dv = (-k*x/m) *dt  # small change in v
    
    x=x+dx # update the variables with small changes
    v=v+dv # update the variables with small changes
    t=t+dt
    
    S+=t,x,v  # stores the calculated values for plotting

    
t,x,v=S.arrays()
```

```python
plot(t,x)
```

```python
plot(t,v)
```

## Parameter exploration

```python
for k in [1,5,10,20]:

    x=3
    v=0
    t=0
    dt=0.01

    #k=1
    m=2


    S=Storage()

    S+=t,x,v

    while t<10:

        dx = (v) *dt  # small change in x
        dv = (-k*x/m) *dt  # small change in v

        x=x+dx # update the variables with small changes
        v=v+dv # update the variables with small changes
        t=t+dt

        S+=t,x,v  # stores the calculated values for plotting


    t,x,v=S.arrays()

    subplot(2,1,1)
    plot(t,x,label=f"k={k}")
    ylabel('x')

    subplot(2,1,2)
    plot(t,v,label=f"k={k}")
    ylabel('v')

    xlabel('Time')
    
legend()
```

```python
count=1
figure(figsize=(20,20))
for k in [1,5,10]:
    for m in [1,5,10]:
        
        x=3
        v=0
        t=0
        dt=0.01

        #k=1
        #m=2


        S=Storage()

        S+=t,x,v

        while t<10:

            dx = (v) *dt  # small change in x
            dv = (-k*x/m) *dt  # small change in v

            x=x+dx # update the variables with small changes
            v=v+dv # update the variables with small changes
            t=t+dt

            S+=t,x,v  # stores the calculated values for plotting


        t,x,v=S.arrays()

        subplot(3,3,count)
        plot(t,x)
        
        title(f"k={k},m={m}")
        ylabel('x')
        xlabel('Time')
        
        count=count+1
    
```

- increasing $k$ the frequency of oscillation goes up
- increasing $m$ decreases the frequency by the same amount -- the frequency depends on the ratio $k/m$ which you can see in the equations
- the period (i.e. time from one peak to another) seems to be about 6 times the ratio of $m/k$.
