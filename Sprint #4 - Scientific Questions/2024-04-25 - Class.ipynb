```python
%matplotlib inline
from pylab import *
from pyndamics3 import Simulation
import pandas as pd
```

```python
data=pd.read_csv('data/influenza1.csv')
data
```

```python
t=array(data['Days'])
y=array(data['Patient 1'])
y = 10**y

t_data=t
y_data=y
```

```python
sim=Simulation()
sim.add("T'=-*T*V",initial_value=1e9,plot=True) # uninfected
sim.add("I'=*T*V-*I",initial_value=.01,plot=True) # infected
sim.add("V'=p*I-c*V",initial_value=0,plot=True)  # viral load
sim.add_data(t=t_data,V=y_data,plot=True)
sim.params(=3.4e-3,=3.4,p=7.9e-3,c=3.3) #(=0.2,=0.1,p=0.5,c=0.1)
sim.run(8)
```

```python
plot(sim.t,log(sim.V))
plot(t_data,log(y_data),'o')
```
