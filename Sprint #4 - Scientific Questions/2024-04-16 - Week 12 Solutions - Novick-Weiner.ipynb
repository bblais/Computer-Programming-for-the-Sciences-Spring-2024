## Novick-Weiner Data

```python
data=pd.read_csv('data/g149novickA.txt',header=None)
data.head()
```

```python
t=data[0]
z=data[1]
plot(t,z,'-o')
```

```python
from pyndamics3 import Simulation
from pyndamics3.fit import fit,Parameter
```

$$
\frac{dz}{dt}= \frac{S}{\tau}-\frac{z}{\tau}
$$

```python
sim=Simulation()
sim.add("z'=S/-z/",0,plot=True)
sim.params(S=1,=2)
sim.add_data(t=t,z=z,plot=True)
sim.run(7)
```

```python
result=fit(sim,
          Parameter("S",value=1,min=0),
          Parameter("",value=1,min=0))
result
```
