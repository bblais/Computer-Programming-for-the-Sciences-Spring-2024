```python
%matplotlib inline
from pylab import *
from sci378 import *
```

```python
y=array([.5,3,2,4.1,.7])  # this is logged data
y=10**y

t=array([0,1,2,3,4])
t_data=t
y_data=y
```

```python
figure(figsize=(6,3))
plot(t,y,'-o')
```

```python
figure(figsize=(6,3))
plot(t,y,'-o')
yscale('log')
```

```python
from pyndamics3 import Simulation
```

```python
sim=Simulation()
sim.add("T'=-*T*V",initial_value=1,plot=True)
sim.add("I'=*T*V-*I",initial_value=1,plot=True)
sim.add("V'=p*I-c*V",initial_value=1,plot=True)
sim.add_data(t=t_data,V=log(y_data),plot=True)
sim.params(=0.2,=0.1,p=0.5,c=0.1)
sim.run(7)
```
